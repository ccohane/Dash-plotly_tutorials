```python
#!usr/lib/bin/activate

import pandas as pd
import numpy as np
import json, jsonify, requests, pprint
```

```python
url = "https://api.overwatchleague.com/stats/matches/10528/maps/1"
```

```python
response=json.loads(requests.get(url).text)
```

```python
print(response['teams'][0]['players'])
```

```python
for x in range(len(response['teams'])):
    for y in range(len(response['teams'][x]['players'])):
        for z in range(len(response['teams'][x]['players'][y]['stats'])):
            print(pd.DataFrame.from_dict(response['teams'][x]['players'][y]['stats'][z],orient='index'))
```
